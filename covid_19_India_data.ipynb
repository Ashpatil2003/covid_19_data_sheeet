## Q.15 Calculate the recovery rate (%) for each State/UT.

```python
newtot=df["New recovered"].sum()
for i,j in df.groupby("Name of State / UT"):
    print(f"{i} ---> {np.ceil((j["New recovered"].sum())/newtot*100)}%")
```

## Q.17 Calculate the monthly total new cases using groupby on month.

```python
df["Date"] = pd.to_datetime(df["Date"])
df["New cases"] = pd.to_numeric(df["New cases"], errors="coerce")
df["Year_Month"] = df["Date"].dt.to_period("M")
monthly_new_cases = df.groupby("Year_Month")["New cases"].sum()
print(monthly_new_cases)
```

## Q.18 Find the daily growth in confirmed cases using diff().

```python
df["Date"] = pd.to_datetime(df["Date"])
df["Total Confirmed cases"] = pd.to_numeric(df["Total Confirmed cases"], errors="coerce")
daily_confirmed = (df.groupby("Date")["Total Confirmed cases"].sum().sort_index())
daily_growth = daily_confirmed.diff()
print(daily_growth)
```
